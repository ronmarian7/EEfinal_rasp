```python
from ultralytics import YOLO
import cv2
import json
import math
```

```python
model = YOLO("../runs/detect/train15/weights/best.pt")
```

```python
with open("../dnn_model/breed/class.json", "r") as f:
    classes = json.load(f)
```

```python
video_path = "../videos/sitting_dog_vid.mp4"
cap = cv2.VideoCapture(video_path)
WindowName = "Webcam"
collor = (255, 0, 0)
FONT_SCALE = 1e-3  # Adjust for larger font size in all images
THICKNESS_SCALE = 1e-3  # Adjust for larger thickness in all images


cv2.namedWindow(WindowName, cv2.WINDOW_NORMAL)
cv2.resizeWindow(WindowName, 1280, 720)
# These two lines will force your "Main View" window to be on top with focus.
cv2.setWindowProperty(WindowName,cv2.WND_PROP_FULLSCREEN,cv2.WINDOW_FULLSCREEN)
cv2.setWindowProperty(WindowName,cv2.WND_PROP_FULLSCREEN,cv2.WINDOW_NORMAL)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model.predict(frame, conf=0.65) #  conf=0.7
        height, width, _ = frame.shape
        font_scale = min(width, height) * FONT_SCALE
        thickness = math.ceil(min(width, height) * THICKNESS_SCALE)
        # Object Detection
        boxes = [box.xyxy.to('cpu').tolist() for box in results[0].boxes]
        class_ids = results[0].boxes.cls.to('cpu').tolist()
        scores = results[0].boxes.conf.to('cpu').tolist()
        for classId, score, box in zip(class_ids, scores, boxes):
            box = box[0]
            (x, y, w, h) = box
            class_name = [key for key, value in classes.items() if value == classId][0]
            cv2.rectangle(frame, (int(x), int(y)), (int(w), int(h)), collor, 2)
            cv2.putText(frame, f"Breed: {class_name} - {score:.3f}", (int(box[0]), int(box[1] - 30)),
                cv2.FONT_HERSHEY_COMPLEX, font_scale, collor, thickness)


        cv2.imshow("Webcam", frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
